**Задание 4**
Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
- оценка 2 и меньше - низкий рейтинг
- оценка 4 и меньше - средний рейтинг
- оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class

```python
import warnings
warnings.filterwarnings('ignore')
movies['average']=0
movies['rating']=''
for id in list(movies['movieId']):
    movies['average'][(movies['movieId']==id)]=rating[(rating['movieId']==id)]['rating'].mean()
    x=rating[(rating['movieId']==id)]['rating'].mean()
    if x<=2: 
        movies['rating'][(movies['movieId']==id)]='низкий рейтинг'
    elif x<=4:     
        movies['rating'][(movies['movieId']==id)]='средний рейтинг'
    elif x>=4:
        movies['rating'][(movies['movieId']==id)]='высокий рейтинг'             
movies[:100]
```
